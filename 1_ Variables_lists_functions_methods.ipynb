The formula is BMI = kg/m2 where kg is a person's weight in kilograms and m2 is their height in metres squared. A BMI of 25.0 or more is overweight, while the healthy range is 18.5 to 24.9. BMI applies to most adults 18-65 years. 

Compute the BMI score of the previous data and you score as well. 

```python
bmi = weight / height ** 2
bmi
```

**Subseting lists**

```python
fam = ["liz", 1.73, "emma", 1.68, "mom", 1.71, "dad", 1.89]
print(fam[1])
#liz is fam[0] !
print(fam[0])
#negative can be used and are particularly useful at the end of the list
print(fam[-1])
print(fam[-2]) #the same as fam[6]
```

**List slicing**

```python
print(fam[3:5]) #contains the 3 rd and 4 th element (starting from index 0)
#So this is like fam([3:5[) in maths !
# We say the start is inclusive and the end is exclusive
#Another trick

print(fam[:4]) # fam[:4] is understood as fam[0:4]
print(fam[0:4])

print(fam[1:]) #Same thing from the other side.
```

<font color='red'>Be very careful when copying and changing lists. Look at this!</font> 

```python
x = ["a", "b", "c"]
y = x
y[1] = "z"
print(y)
print(x)
```

<font color='blue'>Indeed, when we copy a list, we actually copy the reference to the list and not the values inside it. So any update will be feasible for both the list and its copy.</font>    

<font color='red'> Use the Python list() constructor to do it properly!</font> 

```python
x = ["a", "b", "c"]
y = list(x)
y[1] = "z"
print(x)
print(y)
print(type(y))
```

You can also use the slicing operator [:]

```python
x = ["a", "b", "c"]
y = x[:]
y[1] = "z"
print(x)
print(y)
print(type(y))
```
