**Listes de sous-ensembles**

```python
fam = ["liz", 1.73, "emma", 1.68, "mom", 1.71, "dad", 1.89]
print(fam[1])
#liz is fam[0] !
print(fam[0])
#negative can be used and are particularly useful at the end of the list
print(fam[-1])
print(fam[-2]) #the same as fam[6]
```

**List slicing**

```python
print(fam[3:5]) #contains the 3 rd and 4 th element (starting from index 0)
#So this is like fam([3:5[) in maths !
# We say the start is inclusive and the end is exclusive
#Another trick

print(fam[:4]) # fam[:4] is understood as fam[0:4]
print(fam[0:4])

print(fam[1:]) #Same thing from the other side.
```

**Adding and removing elements**

```python
fam_ext = fam + ["me", 1.79]
print(fam_ext)
del(fam[2])
print(fam)
```

<font color='red'>Be very careful when copying and changing lists. Look at this!</font> 

```python
x = ["a", "b", "c"]
y = x
y[1] = "z"
print(y)
print(x)
```

<font color='blue'>Indeed, when we copy a list, we actually copy the reference to the list and not the values inside it. So any update will be feasible for both the list and its copy.</font>    

<font color='red'> Use the Python list() constructor to do it properly!</font> 

```python
x = ["a", "b", "c"]
y = list(x)
y[1] = "z"
print(x)
print(y)
print(type(y))
```

You can also use the slicing operator [:]

```python
x = ["a", "b", "c"]
y = x[:]
y[1] = "z"
print(x)
print(y)
print(type(y))
```

We can think of methods as functions that belong to objects. For instance: 
* <font color='blue'>capitalize() and replace() </font>   are specific methods for the type <font color='blue'> str </font>
* <font color='blue'>bit_length() </font> is a specific method for the type <font color='blue'> int </font>

```python
name='pol'
print(name.capitalize())
print(name.replace('p','P'))
z=15
print(type(z))
z.bit_length()
#If x is nonzero, then x.bit_length() is the unique positive integer k such that 2**(k-1) <= abs(x) < 2**k.
```

```python
fam = ["liz", 1.73, "emma", 1.68, "mom", 1.71, "dad", 1.89]
print(fam.index("liz"))
print(fam.index(1.89))
fam = ["liz", 1.73, "emma", 1.68, "mom", 1.89, "dad", 1.89]
print(fam.index(1.89)) #only the first index
print(fam.count(1.89)) #There are two
print(fam.count("not here")) #0

#append method

fam.append("new")
print(fam)

```
