**Dictionnaries:**

A limitation of lists: 

```python
pop = [30.55, 2.77, 39.21]
countries = ["afghanistan", "albania", "algeria"]
ind_alb = countries.index("albania")
print(ind_alb)
print(pop[ind_alb])
#--------------> not intuitive. We want pop["albania"] but not possible
```

```python
#Dictionnary

world = {"afghanistan" : 30.55 , "albania" : 2.77 , "algeria" : 39.21}
print(world["albania"])
print(type(world)) #dict
```

```python
z = 5
if z % 2 == 0 :
  print("z is even")
else:
  print("z is odd")
```

```python
#loop on numpy arrays
np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
bmi = np_weight / np_height ** 2

for val in bmi :
  print(val)
```

```python
#2d np arrays

import numpy as np
np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
meas = np.array([np_height, np_weight])
for val in meas :
  print(val)
```

```python
#or better
for val in np.nditer(meas):
  print(val)
```
